```python
dataset = pd.read_csv('/Users/dilipkumarvemula/Downloads/Multiple-Linear-Regression-master/50_Startups.csv')
```

```python
dataset.shape
```

```python
X = dataset.iloc[:,:-1]
```

```python
y = dataset.iloc[:,4]
```

```python
dataset
```

```python
state = pd.get_dummies(X['State'],drop_first=True)
```

```python
state
```

```python
X = X.drop('State',axis=1)
```

```python
X = pd.concat([X,state],axis=1)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
linearRegression = LinearRegression()
```

```python
linearRegression.fit(X_train,y_train)
```

```python
y_prid = linearRegression.predict(X_test)
```

```python
from sklearn.metrics import r2_score
```

```python
score = r2_score(y_test,y_prid)
```
